```python
import pandas as pd
df = pd.read_csv('superstore_data.csv')
df.head()
```

```python
df['Income_in_thousands'] = df['Income'] / 100
```

```python
df.drop(['Income'], axis=1, inplace=True)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.boxplot(df['Income_in_thousands'])
plt.title('Income Distribution')
plt.xlabel('Income_in_thousands')
plt.show()
```

```python
%matplotlib inline
```

```python
print(df['Income_in_thousands'].isna().sum())  # Count NaN values
print(df['Income_in_thousands'].dtype) 
```

```python
df['Income_in_thousands'].fillna(df['Income_in_thousands'].mean())
```

```python
plt.hist(df['Income_in_thousands'].dropna(), bins=30)
plt.title('Income Distribution')
plt.xlabel('Income (in thousands)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Calculate IQR (Interquartile Range)

q1 = df['Income_in_thousands'].quantile(0.25) # First quartile (25%)
q2 = df['Income_in_thousands'].quantile(0.50) # Second quartile (50%)
q3 = df['Income_in_thousands'].quantile(0.75) # Third quartile (75%)

# Interquartile Range: Measures the spread of the middle 50% of the data
Iqr = q3 - q1

print ('q1: ', q1)
print ('q2: ', q2)
print ('q3: ', q3)
print ('Iqr: ', Iqr)
```

```python
# Define lower and upper bounds for outliers

lower_bound = q1 - 1.5 * Iqr
upper_bound = q3 + 1.5 * Iqr

print(lower_bound)
print(upper_bound)
```

```python
# Find extreme outliers beyond the upper bound
extreme_outliers = df[df['Income_in_thousands'] > upper_bound]

# Print extreme outlier values
print ('extreme outlier values: ')
print(extreme_outliers[["Id", "Income_in_thousands"]])
```

```python
# Calculate Pearson correlation for Age, Income_in_thousands, and Response
correlations = df[['Age', 'Income_in_thousands', 'Response']].corr()
# Display Pearson correlation
print("Correlation between numerical columns:")
print(correlations)
```

```python
# Calculate Point Biserial correlation for Education and Marital_Status with Response
import scipy.stats as stats

# Label encoding for categorical variables (Education, Marital_Status)
df['Education_encoded'] = df['Education'].astype('category').cat.codes
df['Marital_Status_encoded'] = df['Marital_Status'].astype('category').cat.codes

# Point Biserial correlation for Education and Response
education_corr, _ = stats.pointbiserialr(df['Education_encoded'], df['Response'])
# Point Biserial correlation for Marital_Status and Response
marital_status_corr, _ = stats.pointbiserialr(df['Marital_Status_encoded'], df['Response'])

# Print correlations for categorical variables
print(f"Point Biserial correlation between Education and Response: {education_corr}")
print(f"Point Biserial correlation between Marital_Status and Response: {marital_status_corr}")
```

```python
# Descriptive statistics for customers who responded vs. did not respond
responders = df[df['Response'] == 1]
non_responders = df[df['Response'] == 0]

# Summary statistics for both groups
print("Responders Statistics:")
print(responders.describe())

print("\nNon-Responders Statistics:")
print(non_responders.describe())
```

```python
from scipy import stats
# Compare Means for Numerical Features
# Example: Testing for 'Income_in_thousands' difference
t_stat, p_value = stats.ttest_ind(responders['Income_in_thousands'], non_responders['Income_in_thousands'])
print(f"T-test for Income_in_thousands: t-statistic = {t_stat}, p-value = {p_value}")
```
